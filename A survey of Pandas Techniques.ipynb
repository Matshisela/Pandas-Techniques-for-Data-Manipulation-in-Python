# Pivot Table

The following code shows how a pivot table is created using pandas
https://pandas.pydata.org/docs/reference/api/pandas.pivot_table.html

```python
# Packages
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
# Getting planets data from sns package
df_planets = sns.load_dataset('planets')
df_planets.head(10)
```

```python
#Pivoting by method
pd.pivot_table(df_planets, index= 'method', values = "orbital_period", aggfunc = np.mean)
```

```python
# Calculating sum of Orbital Period
pd.pivot_table(df_planets, index= 'method', columns = "number", values = "orbital_period", aggfunc = np.mean)
```

```python
#Add ing year in the rows
pd.pivot_table(df_planets, index= ['method', 'year'], columns = "number", 
               values = "orbital_period", aggfunc = np.mean)
```

# Plotting
1. https://pandas.pydata.org/docs/reference/api/pandas.plotting.scatter_matrix.html
2. https://gist.github.com/netj/8836201


```python
iris_df = pd.read_csv("https://gist.githubusercontent.com/netj/8836201/raw/6f9306ad21398ea43cba4f7d537619d0e07d5ae3/iris.csv")
iris_df.head(5)
```

```python
# Scatter plot
pd.plotting.scatter_matrix(iris_df[['sepal.length', 'sepal.width', 'petal.length', 'petal.width']], alpha=0.2)
```

```python
# Box Plot
iris_df.boxplot(column=['sepal.length', 'sepal.width', 'petal.length', 'petal.width'])
```

```python
# Andrews Curves
pd.plotting.andrews_curves(iris_df, 'variety')
```

```python
# RadViz Diagram
#RadViz allow to project a N-dimensional data set into a 2D space where the influence of each dimension can be interpreted as a balance between the influence of all dimensions.
pd.plotting.radviz(iris_df, 'variety')
```
